```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
def correlacion_pearson(data_frame, columna1, columna2):
    return np.corrcoef(data_frame[columna1], data_frame[columna2])[0,1];
```

# Datasets
Para los experimentos se crean las siguientes familias de instancias para evaluar los distintos métodos.
1. **beneficio-alto-contagio-bajo**: Son instancias con n = 1..30.
2. **beneficio-bajo-contagio-alto**: Son instancias con n = 1..30.
3. **bt-mejor-caso**: Instancias de mejor caso para el algoritmo de BT con poda, hasta n = 1000
4. **bt-peor-caso**: Instancias de peor caos para el algoritmo de BT con podas.
5. **dinamica**: Instancias con n y M variando para probar el algoritmo de PD, hasta n = 5000
6. **bt-vs-d-mejor**: Instancias para comrpar Backtracking y Prgramación Dinámica cuando M = 2^n, hasta n = 25

# Métodos
Los métodos evaluados en este notebook son los siguientes:
- **FB**: Fuerza Bruta.
- **BT**: Backtracking con podas por optimalidad y factibilidad.
- **BT-F**: Backtracking solamente con poda por factibilidad.
- **BT-B**: Backtracking solamente con poda por optimalidad.
- **DP**: Programación dinámica.

```python
df_resultados = pd.read_csv("resultados/resultado.csv");
df_resultados.head()
```

## Experimento 1: Complejidad de Fuerza Bruta
En este experimento evaluamos el método FB con los datasets bt-peor-caso y bt-mejor-caso. La hipótesis es que FB siempre va a tener un comportamiento exponencial porque tiene peor y mejor caso iguales.

```python
# Graficamos los tiempos de ejecución de cada dataset en función de n.
df_complejidad_FB = df_resultados[(df_resultados["metodo"]=="FB")  & ((df_resultados["dataset"]=="bt-peor-caso") | (df_resultados["dataset"]=="bt-mejor-caso"))].copy();
df_complejidad_FB["tiempo_esperado"] =  2.0 ** df_complejidad_FB["n"] * 0.000160; # Exponencial
fig = sns.scatterplot(data=df_complejidad_FB, x='n', y='tiempo', hue='dataset', alpha =0.6);
fig = sns.lineplot(data=df_complejidad_FB, x='n', y='tiempo_esperado', color='green', label = 'O(2^n)', alpha=0.5);
fig.set(xlabel='n', ylabel='tiempo (ms)');
plt.savefig("graficos/fb-complejidad.svg");
```

```python
# Graficamos los tiempos de ejecución de la serie de instancias X1, ..., Xn vs la complejidad esperada.
df_correlacion_FB = df_resultados[(df_resultados["metodo"]=="FB") & (df_resultados["dataset"]=="bt-mejor-caso")].copy();
df_correlacion_FB["complejidad"] = 2.0 ** df_correlacion_FB["n"];
fig = sns.lmplot(data=df_correlacion_FB, x="tiempo", y="complejidad",line_kws={'color': 'indigo'});
plt.savefig("graficos/fb-correlacion.svg");
print("Índice de correlación de Pearson:", correlacion_pearson(df_correlacion_FB, "tiempo", "complejidad"));
```

## Experimento 2: Complejidad de backtracking para variaciones entre beneficio y contagio
En este experimento evaluamos el método BT con los datasets beneficio-alto-contagio-bajo y beneficio-bajo-contagio-alto. 

```python
# Graficamos los tiempos de ejecución de cada dataset en función de n.
df_complejidad_BT_1 = df_resultados[(df_resultados["metodo"]=="BT")  & ((df_resultados["dataset"]=="beneficio-alto-contagio-bajo") | (df_resultados["dataset"]=="beneficio-bajo-contagio-alto"))].copy();
df_complejidad_BT_1 = df_complejidad_BT_1[df_complejidad_BT_1["tiempo"] < 0.05]
df_complejidad_BT_1["tiempo_esperado"] = df_complejidad_BT_1["n"] * 0.000160; # Lineal
fig = sns.scatterplot(data=df_complejidad_BT_1, x='n', y='tiempo', hue='dataset', alpha =0.6);
fig = sns.lineplot(data=df_complejidad_BT_1, x='n', y='tiempo_esperado', color='green', label = 'O(n)', alpha=0.5);
fig.set(xlabel='n', ylabel='tiempo (ms)');
plt.savefig("graficos/bt-variaciones-1.svg");
```

```python
# Graficamos los tiempos de ejecución de cada dataset en función de n.
df_complejidad_BT_2 = df_resultados[(df_resultados["metodo"]=="BT")  & ((df_resultados["dataset"]=="beneficio-alto-contagio-bajo") | (df_resultados["dataset"]=="beneficio-bajo-contagio-alto"))].copy();
df_complejidad_BT_2["tiempo_esperado"] = np.sqrt(2) ** df_complejidad_BT_2["n"] * 0.000150; # Exponencial
fig = sns.scatterplot(data=df_complejidad_BT_2, x='n', y='tiempo', hue='dataset', alpha =0.6);
fig = sns.lineplot(data=df_complejidad_BT_2, x='n', y='tiempo_esperado', color='green', label = 'O(2^(n/2))', alpha=0.5);
fig.set(xlabel='n', ylabel='tiempo (ms)');
plt.savefig("graficos/bt-variaciones-2.svg");
```

```python
# Graficamos los tiempos de ejecución de la serie de instancias X1, ..., Xn vs la complejidad esperada.
df_correlacion_BT_1 = df_resultados[(df_resultados["metodo"]=="BT") & (df_resultados["dataset"]=="beneficio-alto-contagio-bajo")].copy();
df_correlacion_BT_1["complejidad"] = np.sqrt(2) ** df_correlacion_BT_1["n"];
fig = sns.lmplot(data=df_correlacion_BT_1, x="tiempo", y="complejidad",line_kws={'color': 'indigo'});
plt.savefig("graficos/bt-bacb-correlacion.svg");
print("Índice de correlación de Pearson:", correlacion_pearson(df_correlacion_BT_1, "tiempo", "complejidad"));
```

```python
# Graficamos los tiempos de ejecución de la serie de instancias X1, ..., Xn vs la complejidad esperada.
df_correlacion_BT_2 = df_resultados[(df_resultados["metodo"]=="BT") & (df_resultados["dataset"]=="beneficio-bajo-contagio-alto")].copy();
df_correlacion_BT_2["complejidad"] =  df_correlacion_BT_2["n"];
fig = sns.lmplot(data=df_correlacion_BT_2, x="tiempo", y="complejidad",line_kws={'color': 'indigo'});
plt.savefig("graficos/bt-bbca-correlacion.svg");
print("Índice de correlación de Pearson:", correlacion_pearson(df_correlacion_BT_2, "tiempo", "complejidad"));
```
